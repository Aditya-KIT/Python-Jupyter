# countplot()

```python
# Countplot is a function of seaborn used to count No. of times value appears
```

```python
val=sns.countplot(x='number_courses',data=df)

for container in val.containers:
    val.bar_label(container)
plt.show()
```

# Pie Distribrution

```python
plt.figure(figsize=(4,5)) # Size of Pie chart
course_counts = df['number_courses'].value_counts()

plt.pie(
    course_counts,
    labels=course_counts.index,
    autopct='%1.2f%%'
)
plt.title("Percentage of Number of Courses", fontsize=12)
plt.show()
course_counts
```
