```python
import time

n = 1_000

start = time.perf_counter()

total = 0
for i in range(n):
    total += i * i

end = time.perf_counter()
print(total)
#print("Python loop time:", end - start)
import time

n = 1_000

start = time.perf_counter()

arr = np.arange(n)
total = np.sum(arr * arr)

end = time.perf_counter()
print(total)
#print("NumPy vectorized time:", end - start)
```
