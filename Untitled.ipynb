```python
def interest(amount,duration,rate,paid=0):
    emi= amount-paid
    rate=rate/(12*100)
    duration=duration*12
    emi= (emi*rate*(1+rate)**duration)/((1+rate)**duration-1) 
    print("The Emi is: ")
    return emi
```

```python
a=int(input("Enter the Amount: "))
d=int(input("Enter the Duration: "))
r=int(input("Enter the rate: "))
p=int(input("Enter the Paid: "))
interest(a,d,r,p)
```
