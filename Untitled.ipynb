```python
def interest(amount,duration,rate,paid=0):
    emi= amount-paid
    rate=rate/(12*100)
    duration=duration*12
    emi= (emi*rate*(1+rate)**duration)/((1+rate)**duration-1)
    return emi
```

```python
a=int(input("Enter the Amount: "))
d=int(input("Enter the Duration: "))
r=int(input("Enter the rate: "))
p=int(input("Enter the Paid: "))
emi1= interest(a,d,r,p)
print("The first emi is: ",round(emi1,2))
a=int(input("Enter the Amount: "))
d=int(input("Enter the Duration: "))
r=int(input("Enter the rate: "))
p=int(input("Enter the Paid: "))
emi2= interest(a,d,r,p)
print("The second emi is: ",round(emi2,2))
if(emi1>emi2):
    diff=emi1-emi2
    print("The emi1 is greater than 2 by: ",round(diff,2))
else:
    diff=emi2-emi1
    print("The emi2 is greater than 1 by: ",round(diff,2))
```
